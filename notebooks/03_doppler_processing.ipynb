# 03 Doppler Processing
运动目标的 Range-Doppler 映射（交互式 Plotly 图）。

```python
import os, sys
import numpy as np

sys.path.insert(0, os.path.abspath('..'))
from src.core.radar_params import RadarParams
from src.core.channel_model import Target, simulate_if_signal
from src.processing.range_fft import range_fft
from src.processing.doppler_fft import range_doppler_map
from src.visualization.spectrum_plots import plot_rd_map

params = RadarParams()
target = Target(range_m=40.0, velocity_mps=10.0)
if_data = simulate_if_signal(params, [target])

ranges, rs = range_fft(if_data, params)
vel, rd = range_doppler_map(rs, params)

# ── Interactive Range-Doppler map (Plotly) ────────────────────────────
fig = plot_rd_map(rd, ranges, vel)
fig.update_layout(
    title='Range-Doppler Map (target: 40 m, 10 m/s)',
    width=750, height=450
)
fig.show()

# ── Detected peak (rd shape: [velocity, range]) ───────────────────────
vel_idx, range_idx = np.unravel_index(np.argmax(rd), rd.shape)
print(f'Peak  range: {ranges[range_idx]:.1f} m  (true: {target.range_m:.1f} m)')
print(f'Peak veloc.: {vel[vel_idx]:.1f} m/s (true: {target.velocity_mps:.1f} m/s)')
```
