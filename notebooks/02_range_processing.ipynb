# 02 Range Processing
单目标场景下进行 Range FFT，并绘制距离谱。

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.abspath('..'))
from src.core.radar_params import RadarParams
from src.core.channel_model import Target, simulate_if_signal
from src.processing.range_fft import range_fft
from src.visualization.signal_plots import plot_range_profile, plot_time_waveform

params = RadarParams()
target_range = 30.0  # metres
if_data = simulate_if_signal(params, [Target(range_m=target_range, velocity_mps=0.0)])

# ── Plot 1: IF time-domain signal (first chirp) ────────────────────────
fig1 = plot_time_waveform(if_data[0], params.sample_rate_hz,
                          title='IF Signal – First Chirp (30 m target)')
plt.show()

# ── Plot 2: Range profile (dB) ────────────────────────────────────────
ranges, spec = range_fft(if_data, params)
fig2 = plot_range_profile(ranges, spec, chirp_idx=0)

# Annotate the detected peak
ax = fig2.axes[0]
peak_idx = np.argmax(np.abs(spec[0]))
peak_range = float(ranges[peak_idx])
ax.axvline(peak_range, color='red', lw=1.0, linestyle='--',
           label=f'Peak at {peak_range:.1f} m')
ax.legend()
plt.show()

print(f'Detected range: {peak_range:.2f} m  (true: {target_range:.2f} m)')
```
