# 04 CFAR Detection
1D CA-CFAR 检测示例：信号、阈值与检测结果可视化。

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.abspath('..'))
from src.processing.cfar import ca_cfar_1d, os_cfar_1d

np.random.seed(42)
x = np.random.rand(256) * 0.2
x[120] = 3.0   # single strong target
x[180] = 1.5   # weaker target

# CA-CFAR
thr_ca, det_ca = ca_cfar_1d(x)
# OS-CFAR
thr_os, det_os = os_cfar_1d(x)

# ── Plot: signal, CA-CFAR threshold, OS-CFAR threshold, detections ────
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

for ax, thr, det, label in zip(
        axes, [thr_ca, thr_os], [det_ca, det_os], ['CA-CFAR', 'OS-CFAR']):
    ax.plot(x, lw=1.0, label='Signal', color='steelblue', zorder=2)
    ax.plot(thr, lw=1.2, linestyle='--', label=f'{label} Threshold', color='darkorange', zorder=3)
    ax.scatter(np.where(det)[0], x[det], marker='^', color='red', s=60, zorder=4, label='Detection')
    ax.set_ylabel('Amplitude')
    ax.set_title(label)
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel('Range Bin')
fig.suptitle('1D CFAR Detection', fontsize=13)
fig.tight_layout()
plt.show()

print('CA-CFAR detections at bins:', np.where(det_ca)[0].tolist())
print('OS-CFAR detections at bins:', np.where(det_os)[0].tolist())
```
