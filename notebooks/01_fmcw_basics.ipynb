# 01 FMCW Basics
生成 FMCW Chirp 并观察时域信号与瞬时频率。

```python
import os, sys
import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.abspath('..'))
from src.core.radar_params import RadarParams
from src.core.signal_generator import generate_tx_chirp
from src.visualization.signal_plots import plot_time_waveform

params = RadarParams()
t, tx = generate_tx_chirp(params)

# ── Plot 1: Time-domain waveform (real part of TX chirp) ──────────────
fig = plot_time_waveform(tx, params.sample_rate_hz, title='FMCW TX Chirp – Time Domain')
plt.show()

# ── Plot 2: Instantaneous frequency ───────────────────────────────────
inst_freq_ghz = (params.carrier_freq_hz + params.slope_hz_per_s * t) / 1e9  # GHz
fig2, ax2 = plt.subplots(figsize=(8, 3))
ax2.plot(t * 1e6, inst_freq_ghz, lw=1.2, color='tab:orange')
ax2.set_xlabel('Time (μs)')
ax2.set_ylabel('Instantaneous Frequency (GHz)')
ax2.set_title('FMCW Chirp – Instantaneous Frequency')
ax2.grid(True, alpha=0.3)
fig2.tight_layout()
plt.show()
```
